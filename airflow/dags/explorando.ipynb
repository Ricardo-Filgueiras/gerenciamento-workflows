```python
import os 
import yfinance as yf
import pandas as pd

ticker = "PETR4.SA"
```

```python
petro = yf.Ticker("PETR4.SA")
historico = petro.history(period="1mo")
```

```python
historico['acao_tag'] = ticker
historico.reset_index(inplace=True)
historico 
```

```python
estou_aqui = os.getcwd()
print(estou_aqui)
historico.to_json(f"{estou_aqui[0:-12]}data\\{ticker[0:4]}//{ticker[:4]}.json")
historico.to_csv(f"{estou_aqui[0:-12]}data\\{ticker[0:4]}//{ticker[:4]}.csv")
```
